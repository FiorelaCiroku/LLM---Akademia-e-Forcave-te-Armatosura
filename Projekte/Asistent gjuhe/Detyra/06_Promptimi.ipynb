## 1) Prompt templates (shkruani në kod)
**TODO:** plotësoni 4 template prompts në shqip (instruksionet), me output të kërkuar (format i përgjigjes).


```python
PROMPT_EXPLAIN_PHRASE = """Ti je tutor i anglishtes ushtarake për nxënës shqiptarë.
Shpjego frazën angleze: {phrase}
- Jep përkthimin në shqip
- Shpjego kur përdoret dhe kur nuk përdoret
- Jep një dialog të shkurtër (EN) dhe përkthimin (SQ)
"""

PROMPT_CORRECT = """Ti je tutor i anglishtes. Korrigjo tekstin anglez të nxënësit:
{student_text}
Kthe:
1) Versionin e korrigjuar
2) 2–3 shpjegime në shqip (pse ishte gabim)
3) Një këshillë të shkurtër
"""

PROMPT_ROLEPLAY = """Ti je një partner role-play në anglisht (nivel i thjeshtë).
Skenari: {scenario}
Rregulla:
- Fol në fjali të shkurtra
- Pas çdo përgjigjeje të nxënësit, jep korrigjim në shqip
- Pastaj vazhdo dialogun me një pyetje të re
"""

PROMPT_EMAIL_FEEDBACK = """Ti je tutor shkrimi. Jep feedback për email-in anglez:
{email_text}
Kthe:
- Vlerësim të qartësisë (1–5)
- 3 korrigjime kryesore
- Një version të përmirësuar
"""

print(PROMPT_EXPLAIN_PHRASE.format(phrase="Roger"))
```

## 2) Testim me një pipeline të thjeshtë (opsional)
Nëse modeli lokal s’jep cilësi, s’ka problem — fokus te prompt design.


```python
gen = pipeline("text-generation", model="distilgpt2")
out = gen(PROMPT_EXPLAIN_PHRASE.format(phrase="Roger"), max_new_tokens=80, do_sample=True, top_p=0.9)
print(out[0]["generated_text"][:600])
```
