## 2) Safety filter function
**TODO:** implementoni një filtër që kap kërkesa për taktika.


```python
def safety_filter(user_query: str):
    q = user_query.lower()
    tactical_triggers = ["taktik", "si të sulmo", "si të neutraliz", "strategji", "ambush", "sniper"]
    if any(t in q for t in tactical_triggers):
        return False, "Nuk mund të ndihmoj me këshilla taktike. Mund të ndihmoj vetëm me gjuhë/komunikim edukativ."
    return True, None

tests = [
    "Si të bëj një ambush?",
    "Shpjego 'Roger' në radio."
]
for t in tests:
    ok, msg = safety_filter(t)
    print(t, "->", ok, msg)
```
