## 2) Agent policy
Logjika:
1) Nëse pyetja është 'ku mund ta gjej' -> listim docs
2) Për pyetje për term/frazë -> retrieve + explain
3) Nëse retrieval i dobët -> refuzim


```python
def agent_answer(user_query):
    # Very simple heuristics
    if "ku" in user_query.lower() and "gjej" in user_query.lower():
        return "Kontrollo dokumentet MIL-ENG-001..011 në dosjen documents/."
    hits = tool_retrieve(user_query)
    if not hits:
        return tool_refuse()
    # Extract a candidate phrase from query (toy)
    phrase = None
    for p in ["Roger", "Wilco", "Halt", "SITREP"]:
        if p.lower() in user_query.lower():
            phrase = p
            break
    if phrase:
        return tool_explain_phrase(phrase) + f"\nCitim: {hits[0]['doc']}"
    return f"Përgjigje e bazuar në: {hits[0]['snippet']} (Citim: {hits[0]['doc']})"

print(agent_answer("Çfarë do të thotë Roger?"))
```

## 3) Test me 8 pyetje nga benchmark
