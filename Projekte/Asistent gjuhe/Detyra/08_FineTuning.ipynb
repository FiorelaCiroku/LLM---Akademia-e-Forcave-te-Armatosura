## 1) Ndërtoni dataset instruktesh
Krijoni 30 shembuj të formatit:
{instruction, input, output}
Bazuar në fraza/terma/gabime.
**TODO:** Ruajeni si `train.jsonl`.


```python
import json, random
examples = []
phrases = [
    ("Roger", "Shpjego frazën 'Roger' në radio."),
    ("Wilco", "Shpjego frazën 'Wilco' në radio."),
    ("Halt", "Shpjego frazën 'Halt' në pikë kontrolli.")
]
for phrase, instr in phrases:
    examples.append({
        "instruction": "Ti je tutor i anglishtes ushtarake për nxënës shqiptarë.",
        "input": instr,
        "output": f"(Shembull) '{phrase}' -> përkthim + përdorim + dialog i shkurtër."
    })

# TODO: expand to 30+ examples automatically
while len(examples) < 30:
    examples.append({
        "instruction": "Korrigjo anglishten e nxënësit dhe shpjego në shqip.",
        "input": "We every day patrol the area.",
        "output": "We patrol the area every day. Shpjegim: renditja e fjalëve."
    })

out_path = PROJECT_ROOT / "train.jsonl"
with open(out_path, "w", encoding="utf-8") as f:
    for ex in examples:
        f.write(json.dumps(ex, ensure_ascii=False) + "\n")
print("Wrote:", out_path, "count:", len(examples))
```
